# Linear Layers
## Identity
class torch.nn.Identity(*args, *kwargs)

A placeholder identity operator that is argument-intensive
**Parameters**
* args (Any) : any argument (unused)
* kwargs (Any) : any keyword arguments (unused)

* SHAPE;
*  input (*) , where * means any number or dims
 * output (*)  Same shape as input.

 ### Examples
   

```python
import torch
import torch.nn as nn
```

```python
m = nn.Identity(54, unused_arguments=0.1, unused_arguments_2=False)
input = torch.randn(128, 20)
output = m(input)
print(output.size())
```

### Linear
class torch.nn.Linear(in_features, out_features, bias=True, device=None, dtype=None)

Applies an affine linear transformation to the incoming data : $$ y = xA{^T} + b $$

On certain ROCm devices, when using float16 inputs this module will use different precision for backward.

Parameters:
* in_features (int) - size of the each input sample
* out_features (int) - size of each output sample
* bias (bool) - If set to `False` the layer will not learn an additive bias. Default `False`.

### Shape
* input (*, Hin) where * means any  number of dimensions including none and Hin = in_features
* output (*,Hout) where all but last dim: are the same shape as the inut and Hout = out_features.


### Variables
* weight (torch.Tensor) - the learnable weights of the module of shape
* bias = the learnable bias of the module of shape (out_features).



## Examples

```python
m = nn.Linear(20, 30)
input = torch.randn(128, 20 )
output = m(input)
print(output.size())
```

## Billinear
** class torch.nn.Bilinear(in1_features, in2_features, out_features, bias=True, device=None, dtype=None)**

Applies a bilinear transformation to the incoming data $$ y = x_1^{T} A_x2 + b $$.


## Examples

```python
m= nn.Bilinear(20, 30, 40)
input1 = torch.randn(128, 20 )
input2 = torch.randn(128, 30)
output = m(input1, input2)
print(output.size())
```

## LazyLinear

**  class torch.nn.LazyLinear(out_features, bias=True, device=None, dtype=None)**

In this module, the weight and bias are of torch.nn.UninitializedParameter class. They will be initialized after the first call to forward is done and the module will become a regular torch.nn.Linear module. The in_features argument of the Linear is inferred from the input.shape[-1].
