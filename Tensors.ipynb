Tensors are a specialized data structure that are very similar to arrays and matrices. In PyTorch, we use tensors to encode the inputs and outputs of a model, as well as the model’s parameters.

Tensors are similar to NumPy’s ndarrays, except that tensors can run on GPUs or other specialized hardware to accelerate computing.



```python
import torch
import numpy as np
```

## Tensor Initialization
Tensors can be initialized in various ways. Here is an example.
**Directly from data**
 Tensors can be created directly from the data. The data type is automatically inferred.



```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

**From a Numpy array**
Tensors can be created from NumPy arrays (and vice versa)





```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

**From another tensor**
The new tensor retains the properties (shape, and datatype) of the augmented tensor , unless explicitly overridsen

```python
x_ones = torch.ones_like(x_data) # retain the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor :\n {x_rand} \n")
```

**With random or constact values**

`shape` is a tuple of tenor dimesion. In the functio below , it determines the dimensinality of the output tensor.


```python
shape = (2, 3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor} \n")
```

## Tensor attributes
Tensor attributes describes their shape, datatypes, and the device on which they are stored.


```python
tensor = torch.rand(3, 4)

print(f"Shpae of tensor: {tensor.shape}")
print(f"Datatypes of tensor : {tensor.dtype}")
print(f"Device tensor is stoerd is : {tensor.device}")
```

## Tensor operations
Over 100 tensor operations, including transposing, indexing, slicing, mathematical operations, linear algebra, random sampling, and more
Each of them can be run on the GPU


```python
import torch
import numpy as np

# We move the tensor to the FPU if available
if torch.cuda.is_available():
  tensor = tensor.to("cuda")
  print(f"Device tensor is stored on : {tensor.device}")
```

**Strandard numpy like indexing and slicing**


```python
tensor = torch.ones(4, 4)
# This creates a 4×4 tensor filled with 1s:

tensor[:, 2] = 0
# : means all rows
# 2 means column index 1 (second column)
# = 0 sets that entire column to 0
# So it replaces the second column with zeros.
print(tensor)
```

Joining tensors You can use torch.cat to concatenate a sequence of tensors along a given dimension. See also torch.stack, another tensor joining op that is subtly different from torch.cat.



```python
t1  = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

**Multipling the tensors**


```python
# This computes the element-wise product
print(f"tensor.mul(tensor) \n {tensor.mul(tensor)} \n")
# Alternative syntax
print(f"tensor * tensor \n {tensor * tensor}")
```

**This compute the matrix multipication of the 2 tensors**


```python
print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)} \n")
# Alternative syntax
print(f"tensor @ tensor.T \n {tensor @ tensor.T}")
```

**IN place operations that have a _ sufix are in-place . For eg: `x.copy_y()`, `x.t_()` will  change `x.`**



```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

# Bridge with numpy
Tensors on the CPU and NumPy arrays share their underlying memory locations, and changing one will change the other.


```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

A change in the tensor reflects in the NunPy array


```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

# NumPy array to Tensor

```python
n= np.ones(5)
t = torch.from_numpy(n)
```

change in the NumPy array reflects in the tensor


```python
np.add(n, 1, out=n)
print(f"t : {t}")
print(f"n: {n}")
```
