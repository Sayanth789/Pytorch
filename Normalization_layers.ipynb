# Normalization Layers
## BatchNorm1d
class torch.nn.BatchNorm1d(num_features, eps=1e-0.5, momentum=0.1, affine=True, track_running_stats=True, device=None, dtype=None)

Applies Batch Normalization over a 2D or 3D input

$$
y = \frac{x - \mathbb{E}[x]}{\sqrt{\operatorname{Var}[x] + \varepsilon}} \cdot \gamma + β
$$


The mean and standard-deviation are calculated per-dimension over the mini-batches and γ and β are learnable parameter vectors of size C (where C is the number of features or channels of the input). By default, the elements of γγ are set to 1 and the elements of ββ are set to 0. At train time in the forward pass, the variance is calculated via the biased estimator, equivalent to torch.var(input, unbiased=False). However, the value stored in the moving average of the variance is calculated via the unbiased estimator, equivalent to torch.var(input, unbiased=True).

Also by default, during training the layer keeps running estimates of its computed mean and variance, whcih are then used for normalization during evaluation. The running estimates are lept with a default `momentum` of 0.1.

If `track_running_stats` is set ot `False` this layer then does not keep running estimates, and batfch statistics are instead used during evaluation time as well.


```python
import torch.nn as nn
import torch
```

```python
# Examples
# With the Learnable parameters
m = nn.BatchNorm1d(100)
# Without Learnable Paramters
m = nn.BatchNorm1d(100, affine=False)
input  = torch.randn(20 , 100)
output = m(input)
```

##GroupNorm

 class torch.nn.GroupNorm(num_groups, num_channels, eps=1e-0.5, affine=True, device=None, dtype=None)

 Applies Group Normalization over a mini-batch of inputs

 This layer implements the operation as described in the Paper Group Normalization .


$$
y = \frac{x - \mathbb{E}[x]}{\sqrt{\operatorname{Var}[x] + \varepsilon}} \ * \gamma + β
$$

The input channels are seperated into `num_groups` groups, each containing `num_channels / num_groups` channels. `num_channels` must be divisible by `num_groups`. The mean and stand- deviation are calculated seperately over each group. gamma and beta  are learnable per-channel affine transform parameter vectors  of size `num_channels` if `affine` is `True`. The variance is calculated via the biased estimator equivalent to torch.var(input, unbiased=False)

This layer uses statistics computed from input data in both training and evaluation modes.


```python
#  Examples
input = torch.randn(20, 6, 10, 10)
# Seperate 6 channels into 3 groups
m = nn.GroupNorm(3, 6)
# Seperate 6 channels into 6 groups, (equivalent with InstanceNorm)
m = nn.GroupNorm(6, 6)
# Put all 6 channels into  a single group ( equivalent with LayerNorm)
m = nn.GroupNorm(1, 6)
# Activating the module
output = m(input)

#
```

## LayerNorm
class torch.nn.LayerNorm(normalized_shape, eps=1e-05, elementwise_affine=True, bias=True, device=None, dtype=None)

Applies the Layer Normalization over mini-batch of inputs

This layer implements the operation as described in the paper  Layer Normalization


**Example**

```python
# NLP Example
batch, sentence_length, embedding_dim = 20, 5, 10
embedding = torch.randn(batch, sentence_length, embedding_dim)
layer_norm = nn.LayerNorm(embedding_dim)
# Activate the Module
layer_norm(embedding)

# IMage example
N, C, H, W = 20, 5, 10, 10
input = torch.randn(N, C, H, W)
# Normalize over the last three dimensions (i.e the channel and spatial dimensions)
layer_norm = nn.LayerNorm([C, H, W])
ouput = layer_norm(input)
```

## RMS Norm
Applies the RMS (Root Mean Square Normalization )

clas torch.nn.RMSNorm(normalized_shape, eps=None, elementwise_affine=True, device=None, dtype=None)

###Applies the RMS layer Norm: over a mini batch of inputs.
 $$
y_i = \frac{x_i}{\operatorname{RMS}(x)} \cdot \gamma,
\quad
\operatorname{RMS}(x) = \sqrt{\frac{1}{n} \sum_{i=1}^{n} x_i^2}
$$

The RMS is taken over the last `D` dim: where `D` is the dim: of the `normalized_shape`. For eg: if `normalized_shape` is (3, 5) (a 2-d: shape) the RMS is computed over the last 2 dim: of the input.




```python
# Example
rms_norm = nn.RMSNorm([2, 3])
input = torch.randn(2 , 2, 3)
rms_norm(input)
```
