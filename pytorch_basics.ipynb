```python
# Shapes need to be in right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)


tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

# To perform matrix multiplication, the inner dimensions must match.
# Shape of tensor_A: (3, 2)
# Shape of tensor_B: (3, 2)
# Inner dimensions are 2 and 3, which do not match.

# We need to transpose tensor_B to make the shapes compatible.
# The transpose of a (3, 2) tensor is a (2, 3) tensor.
# After transposing tensor_B, the shapes will be (3, 2) and (2, 3).
# The inner dimensions are now 2 and 2, which match.
print(f"Shape of tensor_A: {tensor_A.shape}")
print(f"Shape of tensor_B: {tensor_B.shape}")

tensor_B_T = tensor_B.T # or tensor_B.transpose(0, 1)
print(f"Shape of tensor_B_T: {tensor_B_T.shape}")

torch.matmul(tensor_A, tensor_B_T)
```

```python
# To correctly do the math we need to take the transpose of the matrix
print(tensor_A)
print(tensor_B.T)
```

```python
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} ( same as above ), tensor_B.T = {tensor_B.shape}\n")

print(f"Multiplying: {tensor_A.shape} * {tensor_B.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"Output shape: { output.shape }")
```

```python
# torch.mm is a shortcut for matmul
torch.mm(tensor_A, tensor_B.T)
```

```python
# Set random seed
torch.manual_seed(1234)

# Check for access to GPU
device = "cuda:" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")


# Create 2 random tensors on GPU

tensor_A = torch.rand(size=(2, 3)).to(device)
tensor_B = torch.rand(size=(2, 3)).to(device)

tensor_A, tensor_B
```

```python
# Performing matmul on the tensors above:
tensor_C = torch.matmul(tensor_A, tensor_B.T)
tensor_C, tensor_C.shape
```

```python
# Find the maximum value of the output in the above
max = torch.max(tensor_C)

# Find the minimum
min = torch.min(tensor_C)

max, min
```

```python
# Find the max and min index value of out: 7
arg_max =  torch.argmax(tensor_C)

# Find the min
arg_min =  torch.argmin(tensor_C)
arg_max, arg_min
```
