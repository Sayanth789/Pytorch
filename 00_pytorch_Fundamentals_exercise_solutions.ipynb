## Create a random tensor with shape (7, 7)


```python
import torch

# Creating a random tensor
X = torch.rand(size=(7, 7))
X, X.shape
```

### Perform a matric multiplication on the tensor above with another random tensor with shape `(1, 7)` (hint: may have to do transpose of the second tensor)

```python
# Creating another tensor
Y = torch.rand(size=(1, 7))

# Z = torch.matmul((X, Y))  # Error as the shape matter
Z = torch.matmul(X, Y.T)

Z, Z.shape
```
