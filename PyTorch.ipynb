#Adaptie MaxPool2D

class torch.nn.AdaptiveMaxPool2d(output_size, return_indices=False)

Applies a 2D adaptive max pooling over an input signal composed of several input planes.
The output is of size Hout x Wout for any input size. The no: of output features is equal to nu: of input planes.


```python
import torch.nn as nn
import torch
```

```python
# Example
# Traget output size of 5 x 7 '
m = nn.AdaptiveMaxPool2d((5, 7))

input = torch.randn(1, 64, 8, 9)
output = m(input)

# Target output size of 7x 8x (square)
m = nn.AdaptiveMaxPool2d((None, 7))
input = torch.randn(1, 64, 10, 9)
output = m(input)
```

## Zeropad 1d
class torch.nn.Zeropad1d(padding)
Pads the input tensor boundaries with zero.

For N-dimensional padding, use `torch.nn.functional.pad()`.

```python
m = nn.ZeroPad1d(2)

input = torch.randn(1, 2, 4)
# This line makes a tensor with 3d: **1, sampling, 2, elements, each having 4 features**

input
```

```python
m(input)
```

```python
# Using different padding for different sides
m = nn.ZeroPad1d((3, 1))
m(input)
```

## nn.ConstantPad2d
Pad input tensor boundaries with a constant value.
class torch.nn.constantPad2d(padding, value)

Pads the input tensor boudaries with a constant value.

for N-D padding Use `torch.nn.functional.pad()`

padding (int, tuple) – the size of the padding. If is int, uses the same padding in all boundaries. If a 4-tuple, uses (padding_lef, tpadding_right, padding_top, padding_botto,)


```python
m = nn.ConstantPad2d(2, 3.5)
input = torch.randn(1, 2, 2)
input

```

```python
m(input)
```

```python
# Using different padding for different sides.
m = nn.ConstantPad2d((3, 0, 2, 1), 3.5)
m(input)
```

##Circularpad1d
class torch.nn.CircularOad1d(padding)
Pads the input tensor using the padding of the input boundaries

Tensor values at the beginning of the dimension are used to pad the end, and values at the end are used to pad the beginning. If negative padding is applied then the ends of the tensor get removed.


```python
m = nn.CircularPad1d(2)
input = torch.arange(8, dtype=torch.float).reshape(1, 2, 4)
input
```

```python
m(input)
```

```python
# Using different paddings for different sides
m = nn.CircularPad1d((3, 1))
m(input)
```

## GELU ( Gaussian Error Linear Unit function).

class torch.nn.GELU(approximate='none')

Applies the Gaussian Error Linear Units function.

GELU(x) = x * phi(x)

Where phi(x) is the Cumulative Distribution  Function for Gaussian DIstribution.

When approximate argument is 'tanh'  Gelu is estimated as
GELU(x) = 0.5 * x * (1 + Tanh(sqrt(2/pi) * (z _ 0.044715* x ^3)))



```python
m= nn.GELU()
input = torch.randn(2)
output = m(input)
```
